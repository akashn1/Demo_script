```python
#long_Term_Column_raw=LongTerm_raw.withColumnRenamed("calendar_year","CALENDAR_YEAR")\
 #       .withColumnRenamed("financial_year","FINANCIAL_YEAR")\
  #      .withColumnRenamed("CDATE_peak","CDATE_PEAK") \
  #      .withColumnRenamed("tsd_peak","TSD_PEAK")\
   #     .withColumnRenamed("tsd_min","TSD_MIN") \
    #    .withColumnRenamed("day_min","DAY_MIN")\
     #   .withColumnRenamed("night_min","NIGHT_MIN")
#print(long_Term_Column_raw.count())
#calendar_year
#financial_year
#ESIWK
#CDATE_peak
#tsd_peak
#tsd_min
#day_min
#night_min

from pyspark.sql.types import DateType,IntegerType
LongTerm_raw_new = LongTerm_raw \
            .withColumn("calendar_year" ,
              LongTerm_raw["calendar_year"]
              .cast(IntegerType()))  \
            .withColumn("financial_year" ,
              LongTerm_raw["financial_year"]
              .cast(IntegerType()))  \
            .withColumn("ESIWK" ,
              LongTerm_raw["ESIWK"]
              .cast(IntegerType()))  \
            .withColumn("CDATE_peak" ,
              LongTerm_raw["CDATE_peak"]
              .cast(DateType()))  \
            .withColumn("tsd_min" ,
              LongTerm_raw["tsd_min"]
              .cast(IntegerType()))  \
            .withColumn("tsd_peak" ,
              LongTerm_raw["tsd_peak"]
              .cast(IntegerType()))  \
            .withColumn("day_min" ,
              LongTerm_raw["day_min"]
              .cast(IntegerType()))  \
            .withColumn("night_min" ,
              LongTerm_raw["night_min"]
              .cast(IntegerType()))  \


        
LongTerm_raw_new.printSchema()
```
